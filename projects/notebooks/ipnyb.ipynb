```python

import geopandas as gpd

# ডেটা লোড
districts = gpd.read_file('bgd_admin_boundaries.shp/bgd_admin2.shp')
rivers = gpd.read_file('bgd_watcrsa_1m_iscgm/bgd_watcrsa_1m_iscgm.shp')

# CRS ঠিক করো
rivers    = rivers.to_crs(epsg=32646)
districts = districts.to_crs(epsg=32646)

# Buffer বানাও
high_risk   = gpd.GeoDataFrame(geometry=rivers.buffer(500),  crs=32646)
medium_risk = gpd.GeoDataFrame(geometry=rivers.buffer(1500), crs=32646)
low_risk    = gpd.GeoDataFrame(geometry=rivers.buffer(3000), crs=32646)

# ✅ .explore() দিয়ে interactive map
m = districts.explore(
    color='lightyellow',
    name='Districts',
    tooltip=False
)

low_risk.explore(
    m=m,                    # আগের map-এর উপর layer যোগ
    color='green',
    opacity=0.3,
    name='Low Risk'
)

medium_risk.explore(
    m=m,
    color='orange',
    opacity=0.4,
    name='Medium Risk'
)

high_risk.explore(
    m=m,
    color='red',
    opacity=0.5,
    name='High Risk'
)

rivers.explore(
    m=m,
    color='blue',
    name='Rivers'
)

# Layer control যোগ করো (চালু/বন্ধ করার বাটন)
folium.LayerControl().add_to(m)

# ফাইলে সেভ করো
m.save('flood_risk_interactive.html')

# Jupyter Notebook-এ সরাসরি দেখাও
m
```
